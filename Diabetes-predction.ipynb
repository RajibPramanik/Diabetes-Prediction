```python
# Import necessary libraries
import numpy as np  # For numerical operations
import pandas as pd  # For data manipulation and analysis

# Import preprocessing, model selection, and evaluation tools from scikit-learn
from sklearn.preprocessing import StandardScaler  # For standardizing data (scaling)
from sklearn.model_selection import train_test_split  # For splitting data into training and testing sets
from sklearn import svm  # Support Vector Machine model
from sklearn.metrics import accuracy_score  # For calculating model accuracy
```

```python
# loading the diabetes dataset to a pandas dataframe
df = pd.read_csv('diabetes.csv')
df
```

```python
# number of rows and Columns in this dataset
df.shape
```

```python
# statistical measures of the data
df.describe()
```

```python
# Count the number of occurrences of each unique value in the 'Outcome' column
df['Outcome'].value_counts()
```

### 0 --> Non-Diabetic

### 1 --> Diabetic

```python
# Group the data by 'Outcome' and calculate the mean for each group
df.groupby('Outcome').mean()
```

```python
# Separate the features (X) and target label (y)
X = df.drop(columns='Outcome', axis=1)  # Drop 'Outcome' column to create feature set
y = df['Outcome']  # Define target variable
```

```python
X
```

```python
y
```

# Data Standardization

```python
# Create an instance of the StandardScaler to normalize the feature values
scaler = StandardScaler()
scaler
```

```python
# Fit the scaler to the feature data (compute the mean and standard deviation for scaling)
scaler.fit(X)
```

```python
# Transform the feature data using the fitted scaler (apply scaling)
standardized_data = scaler.transform(X)
standardized_data
```

```python
# Update X with the standardized data  
X = standardized_data  

# Keep y as the 'Outcome' column  
y = df['Outcome']
```

```python
X
```

```python
y
```

# Train Test Split

```python
# Split the data into training and testing sets  
# test_size = 0.2 -> 20% data for testing, 80% for training  
# stratify = y -> Ensures both train and test sets have a balanced distribution of target labels  
# random_state = 2 -> Ensures reproducibility of the split  
X_train, X_test, y_train, y_test = train_test_split(X, y,  
                                                    test_size=0.2,  
                                                    stratify=y,  
                                                    random_state=2)  

# Display the shape of the full dataset, training set, and testing set  
X.shape, X_train.shape, X_test.shape
```

# Training the Model

```python
# Create an SVM classifier with a linear kernel  
classifier = svm.SVC(kernel='linear')
classifier
```

```python
# Train the Support Vector Machine (SVM) classifier on the training data  
classifier.fit(X_train, y_train)
```

# Model Evaluation & Accuracy Score

```python
# Predict the outcomes for the training data  
X_train_prediction = classifier.predict(X_train)  

# Display the predictions  
X_train_prediction
```

```python
# Calculate the accuracy score on the training data  
training_data_accuracy = accuracy_score(X_train_prediction, y_train)  

# Print the training data accuracy  
print('Accuracy score of the training data:', training_data_accuracy)
```

```python
# Predict the outcomes for the test data  
X_test_prediction = classifier.predict(X_test)  

# Display the predictions  
X_test_prediction
```

```python
# Calculate the accuracy score on the test data  
test_data_accuracy = accuracy_score(X_test_prediction, y_test)  

# Print the test data accuracy  
print('Accuracy score of the test data:', test_data_accuracy)
```

# Making a Predictive System

```python
# Sample input data (features) for prediction  
input_data = (4,154,62,31,284,32.8,0.237,23)  

# Convert the input data to a numpy array  
input_data_as_numpy_array = np.asarray(input_data)  

# Reshape the array to predict for a single instance  
input_data_reshaped = input_data_as_numpy_array.reshape(1, -1)  

# Standardize the input data using the previously fitted scaler  
std_data = scaler.transform(input_data_reshaped)  
print(std_data)  

# Predict the outcome using the trained classifier  
prediction = classifier.predict(std_data)  
print(prediction)  

# Interpret the prediction result  
if prediction[0] == 0:  
    print('The person is not diabetic')  
else:  
    print('The person is diabetic')  
```
